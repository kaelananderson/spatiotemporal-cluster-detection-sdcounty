```python
import geopandas as gpd

# Load shapefiles for each year
gdf_2018 = gpd.read_file('311_Encampment_Reports%2C_City_of_San_Diego%2C_2018/311_Encampment_Reports%2C_City_of_San_Diego%2C_2018.shp')
gdf_2019 = gpd.read_file('311_Encampment_Reports%2C_City_of_San_Diego%2C_2019/311_Encampment_Reports%2C_City_of_San_Diego%2C_2019.shp')
gdf_2020 = gpd.read_file('311_Encampment_Reports%2C_City_of_San_Diego%2C_2020/311_Encampment_Reports%2C_City_of_San_Diego%2C_2020.shp')
gdf_2021 = gpd.read_file('311_Encampment_Reports%2C_City_of_San_Diego%2C_2021/311_Encampment_Reports%2C_City_of_San_Diego%2C_2021.shp')
gdf_2022 = gpd.read_file('311_Encampment_Reports%2C_City_of_San_Diego%2C_2022/311_Encampment_Reports%2C_City_of_San_Diego%2C_2022.shp')
gdf_2023 = gpd.read_file('311_Encampment_Reports%2C_City_of_San_Diego%2C_2023/311_Encampment_Reports%2C_City_of_San_Diego%2C_2023.shp')
```

```python
import pandas as pd
```

```python
gdf_2018['year'] = 2018
gdf_2019['year'] = 2019
gdf_2020['year'] = 2020
gdf_2021['year'] = 2021
gdf_2022['year'] = 2022
gdf_2023['year'] = 2023
```

```python
all_years_gdf = gpd.GeoDataFrame(pd.concat([gdf_2018, gdf_2019, gdf_2020, gdf_2021, gdf_2022, gdf_2023], ignore_index=True))
```

```python
year_tables = {year: all_years_gdf[all_years_gdf['year'] == year] for year in range(2018, 2024)}
```

```python
for year, gdf in year_tables.items():
    gdf.to_csv(f'homeless_camps_{year}.csv', index=False)
```

```python
for year, gdf in year_tables.items():
    gdf.to_file(f'homeless_camps_{year}.shp')
```

```python
# Plot all years in different colors
fig, ax = plt.subplots(figsize=(10, 10))

# Loop through each year's GeoDataFrame and plot it
for year, gdf in year_tables.items():
    gdf = gdf.to_crs(epsg=3857)  # Reproject to Web Mercator
    gdf.plot(ax=ax, marker='o', label=f'{year}', markersize=5)

# Use OpenStreetMap instead of Stamen Toner Lite
ctx.add_basemap(ax, crs=gdf.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik)

plt.legend(title="Year")
plt.title("Homeless Camps in San Diego County (2018-2022)")
plt.show()
```

```python
# Loop through each year and create individual plots
for year, gdf in year_tables.items():
    gdf = gdf.to_crs(epsg=3857)  # Reproject to Web Mercator
    
    # Create a plot for each year
    fig, ax = plt.subplots(figsize=(10, 10))
    
    # Plot the points for the specific year
    gdf.plot(ax=ax, marker='o', color='blue', markersize=5)
    
    # Add OpenStreetMap basemap
    ctx.add_basemap(ax, crs=gdf.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik)
    
    # Set title and display the plot
    plt.title(f"Homeless Camps in San Diego County ({year})")
    plt.show()
```
