```python
import geopandas as gpd
import pandas as pd
```

```python
gdf = gpd.read_file('../data/Public_Transit_Routes%2C_San_Diego_County.geojson')
```

```python
gdf.head()
```

```python
import matplotlib.pyplot as plt

# Plot the data
gdf.plot(figsize=(10, 10), column=None, edgecolor='black', alpha=0.7)
plt.title('Public Transit Routes - San Diego County')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
